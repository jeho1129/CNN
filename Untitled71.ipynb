```python
# Dataset을 load합니다.
import glob
class MyDataset(Dataset):
    def __init__(self, npy_dir):
        self.dir_path = npy_dir
        self.to_tensor = transforms.ToTensor()

        # all npy path
        self.npy_path = glob.glob(os.path.join(npy_dir, '*','*.npy'))

    def __getitem__(self, index):
        # load data
        single_data_path = self.npy_path[index]
        data = np.load(single_data_path, allow_pickle=True)

        image = data[0]
        image = self.to_tensor(image)
        label = data[1]

        return (image, label)

    def __len__(self):
        return len(self.npy_path)

train_data = MyDataset("/content/Font_npy_90_train")
valid_data = MyDataset("/content/Font_npy_90_val")

print(len(train_data))
print(len(valid_data))
```

```python
# define dataloader
batch_size = 50
learning_rate = 0.0003
epochs = 30
train_loader = torch.utils.data.DataLoader(dataset=train_data,
                                           batch_size=batch_size,
                                           shuffle=True)

valid_loader = torch.utils.data.DataLoader(dataset=valid_data,
                                           batch_size=batch_size,
                                           shuffle=True)

# check dataloader
image,label = next(iter(valid_loader))
print(image.shape)
print(label.shape)
```

```python
# device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
!nvidia-smi
```

```python
class CNN(torch.nn.Module):
  def __init__(self):
        super(CNN, self).__init__()
        self.keep_prob = 0.5
        # 1st layer 구성
        self.layer1 = torch.nn.Sequential(
            # Dataset의 in_channel = 1, torch.nn.Conv2d 함수를 통해 (in_channel = 1, out_channel = 16, kernel = 3) 조건에 맞게 layer를 구성한다.
            # ((W - K + 2P) / S) + 1 = ((90 - 3 + 2)/1) + 1 = 90
            # input shape size = 90 * 90이므로, padding = 1로 설정하여 90*90 shape를 유지시킨다.
            torch.nn.Conv2d(1, 16, 3, stride = 1, padding = 1),
            # torch.nn.BatchNorm2d(out_channel) 함수를 이용한다.
            torch.nn.BatchNorm2d(16),
            # Activation function은 ReLu로 설정하여, torch.nn.ReLu 함수를 이용한다.
            torch.nn.ReLU(),
            # Max-polling 수행 후 90*90 -> 45*45으로 변환
            torch.nn.MaxPool2d(kernel_size = 2, stride = 2))
        # 2nd layer 구성
        self.layer2 = torch.nn.Sequential(
            # 1st layer에서 out_channel = 32이므로, torch.nn.Conv2d 함수를 통해 (in_channel = 16, out_channel = 32, kernel = 3) 조건에 맞게 layer를 구성한다.
            # ((W - K + 2P) / S) + 1 = ((45 - 3 + 2)/1) + 1 = 45 -> 45*45
            torch.nn.Conv2d(16, 32, 3, stride = 1, padding = 1),
            # torch.nn.BatchNorm2d(out_channel) 함수를 이용한다.
            torch.nn.BatchNorm2d(32),
            # Activation function은 ReLu로 설정하여, torch.nn.ReLu 함수를 이용한다.
            torch.nn.ReLU(),
            # Max-polling 수행 후 45*45 -> 22*22로 변환
            torch.nn.MaxPool2d(kernel_size = 2, stride = 2))
        # 3rd layer 구성
        self.layer3 = torch.nn.Sequential(
            # 2nd layer에서 out_channel = 4이므로, torch.nn.Conv2d 함수를 통해 (in_channel = 64, out_channel = 128, kernel = 3) 조건에 맞게 layer를 구성한다.
            # ((W - K + 2P) / S) + 1 = ((22 - 3 + 0)/1) + 1 = 20 -> 20*20
            torch.nn.Conv2d(32, 64, 3, stride = 1, padding = 1),
            # torch.nn.BatchNorm2d(out_channel) 함수를 이용한다.
            torch.nn.BatchNorm2d(64),
            # Activation function은 ReLu로 설정하여, torch.nn.ReLu 함수를 이용한다.
            torch.nn.ReLU(),
            # Max-polling 수행 후 22*22 -> 11*11로 변환
            torch.nn.MaxPool2d(kernel_size = 2, stride = 2))
        # 3rd layer 구성
        self.layer4 = torch.nn.Sequential(
            # 2nd layer에서 out_channel = 4이므로, torch.nn.Conv2d 함수를 통해 (in_channel = 64, out_channel = 128, kernel = 3) 조건에 맞게 layer를 구성한다.
            # ((W - K + 2P) / S) + 1 = ((11 - 3 + 0)/1) = 8 -> 8*8
            torch.nn.Conv2d(64, 128, 3),
            # torch.nn.BatchNorm2d(out_channel) 함수를 이용한다.
            torch.nn.BatchNorm2d(128),
            # Activation function은 ReLu로 설정하여, torch.nn.ReLu 함수를 이용한다.
            torch.nn.ReLU(),
            # Max-polling 수행 후 8*8 -> 4*4로 변환
            torch.nn.MaxPool2d(kernel_size = 2, stride = 2))
        # 4th layer 구성
        # 3rd layer에서 max-pooling 후 변환된 10*10 shape, output_channel 64을 input으로 설정하여 10*10*128 -> 120 output로 변환하는 Dense layer를 구성한다.
        self.fc1 = torch.nn.Linear(4 * 4 * 128, 512, bias = True)
        # Error를 줄이기 위한 Weight Initialization 과정을 추가하여 xavier uniform initialization function을 이용한다.
        torch.nn.init.xavier_uniform_(self.fc1.weight)
        self.layer5 = torch.nn.Sequential(
            self.fc1,
            # Activation function은 ReLu로 설정하여, torch.nn.ReLu 함수를 이용한다.
            torch.nn.ReLU(),
            torch.nn.Dropout(p=1 - self.keep_prob))
        # 5th layer 구성
        # 4th layer의 80 output -> 과제 조건의 36 output로 변환하는 Dense layer를 구성한다.
        self.fc2 = torch.nn.Linear(512, 128, bias = True)
        # Error를 줄이기 위한 Weight Initialization 과정을 추가하여 xavier uniform initialization function을 이용한다.
        torch.nn.init.xavier_uniform_(self.fc2.weight)
        self.layer6 = torch.nn.Sequential(
            self.fc2,
            # Activation function은 ReLu로 설정하여, torch.nn.ReLu 함수를 이용한다.
            torch.nn.ReLU(),
            torch.nn.Dropout(p=1 - self.keep_prob))
        # Output layer 구성
        # 5th layer의 36 output -> 10 output로 변환하는 Dense layer를 구성한다.
        self.fc3 = torch.nn.Linear(128, 52, bias=True)
        # Error를 줄이기 위한 Weight Initialization 과정을 추가하여 xavier uniform initialization function을 이용한다.
        torch.nn.init.xavier_uniform_(self.fc3.weight)

  # forward 정의를 이용하여 각 layer를 거쳐 model 학습이 진행되도록 구성한다.
  def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = out.reshape(out.size(0), -1) # flat
        out = self.layer5(out)
        out = self.layer6(out)
        out = self.fc3(out)
        return out
```

```python
# CNN 모델 정의
model = CNN().to(device)
# Cross Entropy Loss를 이용하여 Loss Function 정의
loss_func = torch.nn.CrossEntropyLoss().to(device)
# 과제 조건에 해당하는 Adam Optimizer를 사용하여 Optimizer 정의
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
```

```python
# 모델에서 각 Layer를 통과한 시점에서 Output Shape를 확인.
from torchvision import models
from torchsummary import summary
summary(model, (1, 90, 90))
```

```python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.cross_entropy(output, target)
        loss.backward()
        optimizer.step()
```

```python
def evaluate(model, valid_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in valid_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)

            # 배치 오차를 합산
            test_loss += F.cross_entropy(output, target,
                                         reduction='sum').item()

            # 가장 높은 값을 가진 인덱스가 바로 예측값
            pred = output.max(1, keepdim=True)[1]
            correct += pred.eq(target.view_as(pred)).sum().item()

    test_loss /= len(valid_loader.dataset)
    test_accuracy = 100. * correct / len(valid_loader.dataset)
    return test_loss, test_accuracy
```

```python
for epoch in range(1, epochs + 1):
    train(model, train_loader, optimizer, epoch)
    test_loss, test_accuracy = evaluate(model, valid_loader)

    print('[{}] Test Loss: {:.4f}, Accuracy: {:.2f}%'.format(
          epoch, test_loss, test_accuracy))
```
